```python
df.isnull().sum()
```

```python
df.drop(axis=1, columns=['homepage', 'tagline'], inplace=True)
df
```

```python
def fun_num(col):
    if df[col].isnull().sum()>0:
        df[col].fillna(df[col].mean(), inplace=True)
    print(f"{col}: Mean is {df[col].mean()}")
    print(f"{col}: Median is {df[col].median()}")
```

```python
def fun_cat(col):
    if df[col].isnull().sum()>0:
        df[col].fillna(df[col].mode(), inplace=True)
    print(f"\n There are null values in {col} columns")
    print(f"{col}: Mode of {col} are {df[col].mode()[0]}")
    print(f"{col}: Unique Values in {col} are: {df[col].unique()}")
    print(f"{col}: Number of Missing values in {col} is {df[col].isnull().sum()}")
```

```python
fun_num('runtime')
```

```python
fun_cat('overview')
```

```python
# Task 8 Identify and display the names of all production companies along with they appear in the dataset.
```

```python
all_production_companies = []

for companies_info in df['production_companies']:
    companies = companies_info.split(',')
    for company in companies:
        company = company.strip()      # remove leading and trailing whitespace (spaces, tabs or newline characters) from a string
        if company: 
            all_production_companies.append(company)

unique_production_companies = set(all_production_companies)

print("Names of all production companies:")
for company_name in unique_production_companies:
    print(company_name)
```

```python
print('Interpretation:')
print('These are the production companies of DC Entertainment, Warner Bros, Lightstorm Entertainment, Legendary Pictures, Danjaq, Twentieth century fox film corporation.')
```

```python
# Task 9 Display the names of the top 25 production companies based on the number of movies they have produced in descending order of the number of movies produced.
```

```python
production_company_count = {}

for i in range(len(df)):
    companies_info = df.loc[i, 'production_companies']
    if type(companies_info) == str:
        companies = companies_info.split(',')
        for company in companies:
            company = company.strip()
            if company:
                if company in production_company_count:
                    production_company_count[company] = production_company_count[company] + 1
                else:
                    production_company_count[company] = 1

sorted_production_companies = sorted(production_company_count.items(), key=lambda x: x[1], reverse=True)

print("Top 25 production companies based on the number of movies produced:")
for i in range(min(25, len(sorted_production_companies))):
    company, count = sorted_production_companies[i]
    print(f"{i+1}. {company}: {count} movies")
```

```python
print('Interpretation:')
print('Top 25 production companies based on the number of movies produced has Walt Disney Pictures company has produced 2 movies.')
```

```python
# Task 10 Sort the data in descending order based on reverse and filter the top 500 movies. Find the measures of central tendency for the following columns using the filtered data
# 1. budget
# 2. revenue
# 3. runtime
# Perform outter analysis for the above three columns using box plots
```

```python
sorted_data = df.sort_values(by='revenue', ascending=False)
top_500_movies = sorted_data.head(500)
```

```python
budget_mean = top_500_movies['budget'].mean()
budget_median = top_500_movies['budget'].median()
budget_mode = top_500_movies['budget'].mode().iloc[0]

revenue_mean = top_500_movies['revenue'].mean()
revenue_median = top_500_movies['revenue'].median()
revenue_mode = top_500_movies['revenue'].mode().iloc[0]

runtime_mean = top_500_movies['runtime'].mean()
runtime_median = top_500_movies['runtime'].median()
runtime_mode = top_500_movies['runtime'].mode().iloc[0]

print("Measures of Central Tendency for the Top 500 Movies:")
print("1. Budget:")
print(f"   Mean: {budget_mean}")
print(f"   Median: {budget_median}")
print(f"   Mode: {budget_mode}")
print("\n2. Revenue:")
print(f"   Mean: {revenue_mean}")
print(f"   Median: {revenue_median}")
print(f"   Mode: {revenue_mode}")
print("\n3. Runtime:")
print(f"   Mean: {runtime_mean}")
print(f"   Median: {runtime_median}")
print(f"   Mode: {runtime_mode}")
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
plt.boxplot(top_500_movies['budget'], vert=False)
plt.title('Budget')
plt.subplot(1, 3, 2)
plt.boxplot(top_500_movies['revenue'], vert=False)
plt.title('Revenue')
plt.subplot(1, 3, 3)
plt.boxplot(top_500_movies['runtime'], vert=False)
plt.title('Runtime')
plt.tight_layout()
plt.show()
```
